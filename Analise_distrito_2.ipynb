```python

df['DISTRITO'] = df['DISTRITO'].astype(int)
df['População'] = df['População'].astype(float)

def remover_acentos(texto):
    return ''.join(
        c for c in unicodedata.normalize('NFD', texto)
        if unicodedata.category(c) != 'Mn'
    )
df['BAIRRO 1'] = df['BAIRRO 1'].str.upper().apply(remover_acentos).str.replace(r'\s+', '', regex=True)

bairros_por_distrito = df.groupby('DISTRITO')['BAIRRO 1'].unique()
df['BAIRRO 2'] = df['BAIRRO 2'].astype(str).str.upper().apply(remover_acentos).str.replace(r'\s+', '', regex=True)


populacao_por_bairro = dict(zip(df['BAIRRO 2'], df['População']))

populacao_total_distrito = {}

df['Área (km²)'] = df['Área (km²)'].astype(float)

def remover_acentos(texto):
    return ''.join(
        c for c in unicodedata.normalize('NFD', texto)
        if unicodedata.category(c) != 'Mn'
    )
df['BAIRRO 1'] = df['BAIRRO 1'].str.upper().apply(remover_acentos).str.replace(r'\s+', '', regex=True)

bairros_por_distrito = df.groupby('DISTRITO')['BAIRRO 1'].unique()
df['BAIRRO 2'] = df['BAIRRO 2'].astype(str).str.upper().apply(remover_acentos).str.replace(r'\s+', '', regex=True)


area_por_bairro = dict(zip(df['BAIRRO 2'], df['Área (km²)']))


# Calcule a área total do distrito considera
area_total_distrito = {}

for distrito, bairros in bairros_por_distrito.items():
    bairros_unicos = set(bairros)
    area_total = sum(area_por_bairro.get(bairro, 0) for bairro in bairros_unicos)
    area_total_distrito[distrito] = area_total
    print(f"A area total do distrito {distrito} é: {area_total}")

area_total_geral = sum(area_total_distrito.values())
print(f"\nA area total somada de todos os distritos é: {area_total_geral}")

area_por_bairro_validos = [valor for valor in area_por_bairro.values() if pd.notnull(valor)]
area_total_geral_por_bairro = sum(area_por_bairro_validos)

print(f"\nA área total somada de todos os bairros é: {area_total_geral_por_bairro}")

populacao_total_distrito = {}

for distrito, bairros in bairros_por_distrito.items():
    bairros_unicos = set(bairros)
    populacao_total = sum(populacao_por_bairro.get(bairro, 0) for bairro in bairros_unicos)
    populacao_total_distrito[distrito] = populacao_total

populacao_total_geral = sum(populacao_total_distrito.values())
print(f"\nA população total somada de todos os distritos é: {populacao_total_geral}")


densidade_demografica_distrito = {}
for distrito, populacao_total in populacao_total_distrito.items():
    area_distrito = area_total_distrito[distrito]
    densidade_demografica = populacao_total / area_distrito
    densidade_demografica_distrito[distrito] = densidade_demografica
    print(f"A densidade demográfica do distrito {distrito} é: {densidade_demografica}")

df_densidade_demografica = pd.DataFrame(list(densidade_demografica_distrito.items()), columns=['DISTRITO', 'DENSIDADE DEMOGRÁFICA'])
# Criar um DataFrame para área total por distrito
df_area_distrito = pd.DataFrame(list(area_total_distrito.items()), columns=['DISTRITO', 'ÁREA TOTAL'])

# Criar uma figura e um eixo para o gráfico de área total
fig, ax = plt.subplots(figsize=(10, 6))

# Plotar o gráfico de barras para a área total
barras_area = ax.bar(df_area_distrito['DISTRITO'], df_area_distrito['ÁREA TOTAL'], color='lightblue')

# Adicionar os valores no topo de cada barra
for barra in barras_area:
    altura = barra.get_height()
    ax.text(barra.get_x() + barra.get_width() / 2, altura, f'{altura:.2f}',
            ha='center', va='bottom', fontsize=9)

# Adicionar título e rótulos ao gráfico de área
ax.set_title('Total Area by District')
ax.set_xlabel('Districts\n\nFigure 8', labelpad=10)
ax.set_ylabel('Total Area (km²)')
ax.set_xticks(df_area_distrito['DISTRITO'])
ax.set_xticklabels(df_area_distrito['DISTRITO'], rotation=45, ha='right')

# Mostrar o gráfico
plt.tight_layout()  # Ajustar layout para evitar cortes
plt.show()
```

# **IDH Geral**

### 1. **Coeficiente Angular (\( \beta \))**:
- O valor de \( \beta \), que é o coeficiente angular da linha de regressão, foi calculado como **-0.21**. Isso indica uma **correlação negativa** entre a população e o IDH Geral, ou seja, conforme a população aumenta, o IDH Geral tende a diminuir levemente.
- O valor negativo de \( \beta \) sugere que nos bairros com populações maiores, o índice de desenvolvimento humano (IDH) geral tende a ser um pouco menor, mas o impacto dessa diminuição é relativamente pequeno, considerando o valor de \( \beta \).

### 2. **Coeficiente de Determinação (\( R^2 \))**:
- O coeficiente de determinação \( R^2 \) foi calculado como **0.04**, o que indica que apenas **4% da variação** do IDH Geral pode ser explicada pela variação da população.
- Isso significa que a relação entre população e IDH Geral é **muito fraca**. Ou seja, a população não é um bom preditor para o IDH Geral com base nesses dados, e a maior parte da variação no IDH Geral é influenciada por outros fatores que não estão sendo considerados nesse gráfico.
